```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def derivative(func,var):
    return sp.diff(func,var)
```

```python
x=sp.symbols('x')
y=(x+3)**2
gradient=derivative(y,x)
```

```python
x0=2
lr=0.1
epochs=50
```

```python
def gradient_descent(func,grad,x_start,lr,epochs):
    minima_x=x_start

    x_history=[]
    y_history=[]

    print(f"Initial values:x={minima_x},y={func.subs(x,minima_x)}")

    for i in range(epochs):
         gradVal=grad.subs(x,minima_x)
         minima_x=minima_x-lr*gradVal
         x_history.append(minima_x)
         y_history.append(func.subs(x,minima_x))
    return minima_x,func.subs(x,minima_x),x_history,y_history
```

```python
x_minima,y_minima,x_history,y_history=gradient_descent(y,gradient,x0,lr,epochs)
print(f"local minima found at : x={x_minima},y={y_minima}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x_history,y_history)
plt.plot(x_history,y_history)
plt.scatter(x_history[-1],y_history[-1],c='Green')
plt.title('Gradient Descent')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
